## Inheritance 

Inheritance is much powerful than aggregation. Concept wise it is same (relationship). 

Inheritance allows us to define a class that inherits all the methods and properties from another class. Parent class is the class being inherited from, also called base class. Child class is the class that inherits from another class, also called derived class.

- non private attribute
- non private methods

```python
class User: # parent class
    def __init__(self):
        self.name = "Saurabh"
        self.gender = "Male"

    def login(self):
        print("Login Done!")

    def register(self):
        print("register Done!")

class Student(User):
    def __init__(self):
        self.rollno = 12

    def enroll(self):
        print("Enrolled the course!")


class Instructor(User):
    def __init__(self):
        self.idno = 120

    def create(self):
        print("created the course!")
```

```python
s = Student()
```

```python
s.enroll()
```

```python
s.login()
```

```python
s.register()
```

```python
i = Instructor()
i.create()
i.login()
```
