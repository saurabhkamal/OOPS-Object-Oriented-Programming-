### Types of Inheritence in OOP

- Single Inheritance
- Multilevel Inheritance
- Hierarchical Inheritance
- Multiple Inheritance

```python
# Single Inheritance
class User:  # Parent Class
    def __init__(self):
        self.name = "Saurabh"
        self.gender = "Male"

    def login(self):
        print("Login Done!")

    def register(self):
        print("Register Done!")

class Student(User):  # Child class
    def __init__(self):
        self.rollno = 12
    
    def enroll(self):
        print("Enrolled the course")
```






Hybrid Inheritance: Combination of every inheritance, but very less used

```python
# Multilevel Inheritance

# GrandFather
class Product:
    def review(self):
        print("Product Customer Review")

# Father
class Phone(Product):
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a Phone")


# Child
class SmartPhone(Phone):
    pass

S = SmartPhone(200000, "Samsung", 3)
S.buy()
S.review()
```

```python
# Hierarchical

# Parent
class Phone:
    def __init__(self, price, brand, camera):
        print("Inside Phone Constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("Buying a Phone")

# child
class SmartPhone(Phone):
    pass

# Child
class FeaturePhone(Phone):
    pass

S = SmartPhone(100000, "Apple", "3")
S.buy()
```

```python
F = FeaturePhone(100000, "Apple", "3")
F.buy()
```

```python
# Multiple
# parent
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone constructor")
    self.__price = price
    self.brand = brand
    self.camera = camera

  def buy(self):
    print("Buying a Phone")

# Parent
class Product:
  
  def __init__(self):
    pass
  
  def review(self):
    print("Product Customer review")


#child
class SmartPhone(Phone, Product):
  pass

S = SmartPhone(100000, "apple", "3")
S.buy()
S.review()
```
